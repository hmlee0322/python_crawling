```python
blank = []
for i in range (4100,4110):
    url='http://gall.dcinside.com/board/view/?id=onepiece&no='+str(i)
    li = []
    print(i)
    j=requests.get(url)
    bs5=BS(j.text)
    T=bs5.select('.wt_subject dd')[0].text
    N=bs5.select('.user_layer span')[0].text
    num=bs5.select('.dd_num')[0].text.strip()
    gul=bs5.select('.s_write td')[0].text
    li.append(T)
    li.append(N)
    li.append(num)
    li.append(gul)
    blank.append(li)

```

```python
##리스트의 리스트 형태로 만들고 csv로 저장
header=['제목','아이디','조회수','내용']
file = open(str('abc') + str('.csv'), 'w')
writer=csv.writer(file,delimiter=',',lineterminator = '\n')
writer.writerow(header)
for j in blank:
    print(j)
    writer.writerow(j)
file.close()
```
