```python
import requests
from bs4 import BeautifulSoup as bs
import pandas
from pandas import DataFrame, Series
```

```python
res = requests.get('http://www.botoacai.com/')
res.encoding = 'euc-kr'
```

```python
if res.status_code == 200:
    soup = bs(res.text, 'html5lib')
    
    items = soup.select('#mk_center > table > tbody > tr > td > table > tbody > tr > td')
    
    table = []    
    
    for item in items:
        row = []

        #제품명 리스트 저장
        prod_name = item.select('b') 
        #최저 가격 = price_1 저장
        price_1 = item.findAll('div', {'style':'padding-bottom:3px; color:#000;'})
        #소비자가 = price_2 저장
        price_2 = item.select('strike')
        events = str(item).split('<br/>')

        row.append(prod_name[0].text.strip() if len(prod_name) > 0 else "")
        row.append(price_1[0].text.strip() if len(price_1) > 0 else "")
        row.append(price_2[0].text.strip() if len(price_2) > 0 else "")
        row.append(events[1] if len(events) == 3 else "")
        
        table.append(row)

table
            
```

```python
items_df = DataFrame(table)
items_df.columns = ['제품명','실판매가','소비자가','상품구성']
items_df.to_csv('./보뚜슈퍼푸드_list.csv')
```
