```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
res = requests.get('http://emart.ssg.com/planshop/detail.ssg?dispCmptId=6000077122')
```

```python
res.text
```

```python
if res.status_code == 200:
    soup = bs(res.text, 'html.parser')
    
    items = soup.select('#content > div.plan_wrap > table')
    
    table = []
    for item in items:
        
        prod_name = item.findAll('span', {'class':'tx'})
        price = item.findAll('div', {'class':'price'})
        
        for i in range(len(prod_name)):
            row = []
            row.append(prod_name[i].text.strip())
            row.append(price[i].text.strip())
            
            table.append(row)

table
```

```python
import pandas
from pandas import DataFrame, Series
```

```python
items_df = DataFrame(table)
items_df.columns = ['제품명','가격']
items_df.to_csv('./노브랜드제품_list.csv')
```
