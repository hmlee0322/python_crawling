```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
res = requests.get("https://www.naver.com/")
res.status_code
```

```python
soup = bs(res.content, 'html.parser')
soup
```

```python
ranking_elements = soup.select('.ah_item')
```

```python
table = []

for ranking_element in ranking_elements:
    row = []
    ranking_num = ranking_element.select('.ah_r')
    ranking_title = ranking_element.select('.ah_k')
    
    row.append(ranking_num[0].text.strip() if len(ranking_num[0]) > 0 else "")
    row.append(ranking_title[0].text.strip() if len(ranking_title[0]) > 0 else "")
    
    table.append(row)
```

```python
import pandas
from pandas import DataFrame
```

```python
df_realtime_ranking = DataFrame(table)
df_realtime_ranking.columns = ['순위', '실시간검색어']
df_realtime_ranking
```
