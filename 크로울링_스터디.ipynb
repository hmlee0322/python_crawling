```python
import requests
from bs4 import BeautifulSoup

def get_single_item_data(item_url):
    source_code = requests.get(item_url)
    plain_text = source_code.text
    soup = BeautifulSoup(plain_text)
    # if you want to gather information from that page
    for item_name in soup.findAll('span', {'class': 'pluginCountTextDisconnected'}):
        try:
                print(item_name.string)
        except:
                print("error")

get_single_item_data("http://news.sky.com/story/1330046/are-putins-little-green-men-back-in-ukraine")
```

```python
###### 크로울링 실전 #######
#원피스 갤러리의 첫번째 페이지 글제목을 가져오자
k=[]
for i in range(29):
    url='http://gall.dcinside.com/board/lists/?id=onepiece'
    j=requests.get(url)
    bs5=BS(j.text)
    T=bs5.select('.t_subject')[i+2].text
    #if '[' in T:
    #    T=T.split('[')[0]
    k.append(T)
```

```python
k
```

```python
###### 크로울링 실전 #######
##원피스 갤러리의 글 제목을 10페이지 까지 긁어와 보자
k=[]
for i in range(10):
    baseurl='http://gall.dcinside.com/board/lists/?id=onepiece'
    fullurl=baseurl+'&page='+str(i)
    j=requests.get(fullurl)
    bs5=BS(j.text)
    for i in range(len(bs5.select('.t_subject'))-2):
        T=bs5.select('.t_subject')[i+2].text
        #if '[' in T:
        #    T=T.split('[')[0]
        k.append(T)
```

```python
### 크로울링 실전
### 원피스 갤러리 10페이지까지 글제목, 내용, 작성자, 조회수를 긁어오자
blank=[]
s=2892
for i in range(1):
    baseurl='http://gall.dcinside.com/board/view/?id=onepiece'
    url=baseurl+'&page='+str(i+1)
    
    for k in range(29):
        try:
            t=[]    
            fullurl=url+'&no='+str(s-k)
            print(fullurl)
            j=requests.get(fullurl)
            aaa=BS(j.text)
            t.append(aaa.select('.wt_subject dd')[0].text)
            t.append(aaa.select('.user_layer')[0].text)
            t.append(aaa.select('.dd_num')[0].text.strip())
            t.append(mod(aaa.select('.s_write tr')[0].text))
            blank.append(t)
        except:
            pass
    s=s-29
```

```python
#못읽는 문자 지우기
def mod(cha):
    cha=cha.replace('\xa0','')
    return cha
```

```python
##리스트의 리스트 형태로 만들고 csv로 저장
header=['제목','아이디','조회수','내용']
file = open(str('원피스6') + str('.csv'), 'w')
writer=csv.writer(file,delimiter=',',lineterminator = '\n')
writer.writerow(header)
for j in blank:
    print(j)
    writer.writerow(j)
file.close()
```
